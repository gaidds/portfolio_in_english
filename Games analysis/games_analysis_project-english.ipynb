The Strimchik online store sells computer games all over the world.

Historical data up to 2016 is available from open sources on:

### Data preprocessing

#### Replacing column names

We bring the names of the columns of the table to lower case:

```python
games.columns = games.columns.str.lower()
games.head()
```

#### Replacing data types

Let's look at the values ​​of `user_score` :

```python
games["user_score"].value_counts()
```

Let's see how many missing values ​​it has:

```python
# let's see the missing values
print("Number of passes:", games["user_score"].isna().sum())
```

Note that in `user_score` there are values ​​`tbd` (“to be determined”), which also mean that the values ​​have not yet been determined. Let's replace `tbd` with NaN and change the data type of the column to `float64` :

```python
games["user_score"] = games["user_score"].replace("tbd", np.nan)
# checking the missing values again
print("Current number of passes:", games["user_score"].isna().sum())

# changing the data type of the column to float64
games["user_score"] = pd.to_numeric(games["user_score"])
```

Change the data type for `year_of_release` to `Int64` :

```python
games["year_of_release"] = games["year_of_release"].astype("Int64")
games.info()
```

#### Pass handling

Let's look at the pass data:

```python
games.isna().sum()
```

Let's remove the lines where there are gaps in `name`, `genre`, `year_of_release` (since there are quite a few of them in the total number of lines):

```python
games = games.dropna(subset=["name", "genre", "year_of_release"]).reset_index(drop=True)
```

Let's check:

```python
games.isna().sum()
```

Since the ESRB rating is assigned only for games in America and Canada, we will replace the remaining gaps with `no_rating` for games in other regions:

```python
games["rating"] = games["rating"].fillna("no_rating")
```

Let's check:

```python
games.isna().sum()
```

### Hypothesis testing

#### Average user ratings for Xbox One and PC platforms are the same

Let's test the hypothesis about the equality of the user ratings of the Xbox One and PC platforms.

The user ratings of the Xbox One and PC platforms are populations independent of each other.
When comparing the means of two populations, the null hypothesis can be formulated as: "the means of the populations are equal."
And the alternative - that they are not equal.

Let's use the t-test: `scipy.stats.ttest_ind(array1, array2, equal_var, alternative)`.

t-test works for any sample size

**Null hypothesis** - average user ratings of Xbox One and PC platforms are the same

**Alternative hypothesis** - average user ratings of Xbox One and PC platforms are different

Let's do a two-tailed `t-test` with a significance level of `alpha = 0.01`

```python
alpha = 0.01  # level of statistical significance

results = st.ttest_ind(
    games_new.loc[games["platform"] == "XOne"]["user_score"],
    games_new.loc[games["platform"] == "PC"]["user_score"],
    equal_var=False,
    nan_policy="omit",
)

print(f"p-value : {round(results.pvalue, 2)}")

# checking p-value
if results.pvalue < alpha:
    print("Rejecting the null hypothesis")
else:
    print("No reason to reject the null hypothesis")
```
